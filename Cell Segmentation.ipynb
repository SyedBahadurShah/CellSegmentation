```python
# Importing All Packages for Cell Segmentation
import cv2, numpy as np, matplotlib.pyplot as plt
```

```python
cellImage = cv2.imread("Input/Sample #1.png") # Reading Image Input Samples
copyCellImage = cellImage.copy() # For Future Usage in Code
plt.imshow(cellImage)
```

```python
grayCellImage = cv2.cvtColor(cellImage, cv2.COLOR_BGR2GRAY) # Converting BGR Image into Gray Scale
plt.imshow(grayCellImage)
```

```python
medianBlurCellImage = cv2.medianBlur(grayCellImage, 5) # Applying Median Blur on Gray Scale Cell Image
plt.imshow(medianBlurCellImage)
```

```python
# Delecting & drawing circles around the cells putting blue dot at the center of each cell
houghCirclesCellImage = cv2.HoughCircles(medianBlurCellImage, cv2.HOUGH_GRADIENT, 1, 120, param1 = 100, param2 = 30, minRadius = 0, maxRadius = 0)
circledCellImageData = np.uint16(np.around(houghCirclesCellImage))

for i in circledCellImageData[0, :]:
    cv2.circle(cellImage, (i[0], i[1]),i[2], (0,255,0), 6)
    cv2.circle(cellImage, (i[0], i[1]), 2, (0,0, 255), 20)
plt.imshow( cellImage )
```

```python
sampleCounter = 1
for x in circledCellImageData[0, :]:
    croppedCellImage = copyCellImage[x[1] - x[2] : x[1] + x[2], x[0] - x[2] : x[0] + x[2]]
    try: 
        cv2.imwrite("Output/Sample #" + str(sampleCounter) + ".jpg", croppedCellImage)
        print("Cell Sample # " + str(sampleCounter) + " Saved")
        sampleCounter = sampleCounter + 1        
    except:
        continue
```
